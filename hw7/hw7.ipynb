```python
# Homework 7

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates
from scipy.stats import skew
from matplotlib.offsetbox import AnchoredText
from datetime import timedelta
from scipy.stats import t
import os
from scipy.stats import norm
from scipy.signal import welch
from scipy.stats import chi2
from sklearn.linear_model import LinearRegression


file_loc = '/Users/kboot/phy577_projects/hw7/data/FPS_UVP.csv'

data = pd.read_csv(file_loc)

print(data)
```

```python
# 3. Construct two data matrices (be sure to remove the mean & trend from each column)

# d=[u, v, p], and, w=[u,v].

from sklearn.linear_model import LinearRegression

u =np.asarray(data['EastVelocity'])
v =np.asarray(data['NorthVelocity'])
p =np.asarray(data['Pressure'])

mean_d = []
mean_w = []

d = np.column_stack((u,v,p))
w = np.column_stack((u,v))

for i in range(d.shape[1]):
    
    #Calculate and remove mean from each column
    
    mean_d.append(d[:, i].mean())
    d[:, i] -= mean_d[i]
    
    # Linear fit and remove trend from each column
    
    model = LinearRegression().fit(np.arange(len(d)).reshape(-1, 1), d[:, i])
    trend = model.predict(np.arange(len(d)).reshape(-1, 1))
    d[:, i] -= trend

    
for i in range(w.shape[1]):
    mean_w.append(w[:, i].mean())
    w[:, i] -= mean_w[i]
    model = LinearRegression().fit(np.arange(len(w)).reshape(-1, 1), w[:, i])
    trend = model.predict(np.arange(len(w)).reshape(-1, 1))
    w[:, i] -= trend
    
#print('Means:', means)
#print(d), print(w)

# 4. Estimate the covariance matrix of d

# C = dt*d/N

N = len(d)

cov_matrix = np.dot(d.T, d) / N

#print(cov_matrix.shape)

cov_mat_names = [
    ['s2u', 's2uv', 's2up'],
    ['s2vu', 's2v', 's2vp'],
    ['s2pu', 's2pv', 's2p']
]

#print(cov_mat_names)

# 5.Using these values, what are the zero-lagged correlation coefficients ρu,v, ρp,u, ρp,v? 
# What does the sign and magnitude of ρ tell you about their relationship, i.e., covariance? 

'''
# Assuming cov_matrix contains the covariance matrix
s2u = cov_matrix[0, 0]
s2v = cov_matrix[1, 1]
s2p = cov_matrix[2, 2]
s2u_v = cov_matrix[0, 1]
s2u_p = cov_matrix[0, 2]
s2v_p = cov_matrix[1, 2]

# Calculate zero-lagged correlation coefficients
rho_u_v = s2u_v / np.sqrt(s2u * s2v)
rho_p_u = s2u_p / np.sqrt(s2p * s2u)
rho_p_v = s2v_p / np.sqrt(s2p * s2v)

print(f"Correlation coefficients: ")
print(f"ρu,v: {round(rho_u_v,3)}")
print(f"ρp,u: {round(rho_p_u, 3)}")
print(f"ρp,v: {round(rho_p_v, 3)}")
'''

##################### Alternate method? ############################

# Assuming cov_matrix contains the covariance matrix
# Define the correlation matrix to store correlation coefficients
correlation_matrix = np.zeros_like(cov_matrix)

# Calculate the diagonal elements (variances) from the covariance matrix
variances = np.sqrt(np.diag(cov_matrix))

# Calculate the correlation coefficients using the formula: rho = cov(X, Y) / (std(X) * std(Y))
for i in range(cov_matrix.shape[0]):
    for j in range(cov_matrix.shape[1]):
        correlation_matrix[i, j] = cov_matrix[i, j] / (variances[i] * variances[j])

correlation_matrix = np.round(correlation_matrix, 3)

corr_mat_names = [
    ['pu', 'puv', 'pup'],
    ['pvu', 'pv', 'pvp'],
    ['ppu', 'ppv', 'pp']
]
print("Zero-lagged correlation coefficients:", correlation_matrix)


# 6. Estimate the singular value decomposition on both matrices, e.g., d = (U S)ΦT using the Python equivalent of use the command svd(d,0) so that S is square.
#### np.linalg.svd is the equivalent function in Python

# SVD on the correlation matrix
U_corr, S_corr, V_corr_transpose = np.linalg.svd(correlation_matrix, full_matrices=False)

# SVD on the covariance matrix
U_cov, S_cov, V_cov_transpose = np.linalg.svd(cov_matrix, full_matrices=False)

# Make S matrix square
S_corr_square = np.diag(S_corr).round(3)
S_cov_square = np.diag(S_cov).round(3)
'''
print("SVD on the correlation matrix:")
print("U_corr:")
print(U_corr.round(3))
print("S_corr:")
print(S_corr_square)
print("V_corr_transpose:")
print(V_corr_transpose.round(3))

print("\nSVD on the covariance matrix:")
print("U_cov:")
print(U_cov.round(3))
print("S_cov:")
print(S_cov_square)
print("V_cov_transpose:")
print(V_cov_transpose.round(3))
'''

# 7.Estimate the eigen-values λk = σk2 and uncertainties ∆λk of d. 
# How many effective degrees of freedom do you have? 
# Plot the fraction of total variance described by each mode with error-bars. 
# Are the eigen-values/vectors distinct? 

# Calculate the uncertainties in the singular values of d matrix

U, S, VT = np.linalg.svd(d, full_matrices=False)
lambda_k = S ** 2 # eigenvalues
N = len(d)  
delta_sigma_k = np.sqrt(2) / np.sqrt(N)

# Calculate uncertainties in eigenvalues
delta_lambda_k = 2 * S * delta_sigma_k  # Twice the uncertainties in singular values

effective_dof = np.sum(S ** 2) / (S[0] ** 2)

print("Uncertainties in the singular values for 'd' matrix:", round(delta_sigma_k,3))
print("Eigenvalues: ", lambda_k.round(3))
print("Uncertainties in eigenvalues: ", delta_lambda_k.round(3))
print('Effective DOF:', effective_dof.round())

modes = VT.T  

dot_products = np.dot(modes[:3], modes[:3].T)  
print("Dot Products (Inner Products) between the Modes:\n", dot_products.round())

# Calculate the magnitude of each mode
magnitudes = np.linalg.norm(modes[:3], axis=1)  
print("Magnitudes of the First Three Modes:", magnitudes)
```
