```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylib import *
from datetime import datetime
import matplotlib.dates as mdates
from scipy.stats import skew
from matplotlib.offsetbox import AnchoredText

#!pip install folium
#!pip install geopy

import folium
from geopy.distance import geodesic

# Santa Barbara, CA - Station ID: 9411340 ----- 0
# Monterey, CA - Station ID: 9413450 ------- 1
# metric, GMT, MSL, csv, 6-minute intervals

file_loc = ['/Users/kboothomefolder/phy577_projects/hw4/data/CO-OPS__9411340__wl.csv','/Users/kboothomefolder/phy577_projects/hw4/data/CO-OPS__9413450__wl.csv']
stas = [pd.read_csv(file) for file in file_loc]

stas[0], stas[1] = pd.DataFrame(stas[0].iloc[:, :-5]), pd.DataFrame(stas[1].iloc[:, :-5])

################# Create Dataframe ##########################
#print(stas[0].columns)

# Convert "Date Time" column to numpy datetime64 array

for i in range(len(stas)):
    stas[i]['datestr'] = np.asarray(pd.to_datetime(stas[i]['Date Time']))
    stas[i]['msl'] = np.asarray(pd.to_numeric(stas[i][' Water Level']))
    stas[i]['sigma'] = np.asarray(pd.to_numeric(stas[i][' Sigma']))
    stas[i].drop(stas[i].columns[0:3], axis=1, inplace=True)
    
################# Plot Locations on Map w/ distance ##########################

SB_coords = (34.405 , -119.6917)
M_coords = (36.6002, -121.8947)

# Calculate distance between the two locations
dist = geodesic(SB_coords, M_coords).miles

# Create a folium map centered between the two locations
m = folium.Map(location=[(SB_coords[0] + M_coords[0]) / 2, (SB_coords[1] + M_coords[1]) / 2], zoom_start=6)

# Add markers for Santa Barbara and Monterey
folium.Marker(SB_coords, tooltip="Santa Barbara, CA").add_to(m)
folium.Marker(M_coords, tooltip="Monterey, CA").add_to(m)

# Create a line between the two locations to show the distance
line = folium.PolyLine([SB_coords, M_coords], color='blue', weight=2, opacity=1)
line.add_to(m)

# Display the distance as a tooltip on the line
line.add_child(folium.Tooltip(f"Distance: {dist:.2f} miles"))

output_dir = '/Users/kboothomefolder/phy577_projects/hw4/figures'
m.save(os.path.join(output_dir, 'SB_M_map.html'))

print("Distance from Santa Barbara to Monterey, CA:", round(dist, 2), "miles")
################# Plot MSL ##########################

plt.figure(figsize=(10, 8))
plt.plot(stas[0]['datestr'].values, stas[0]['msl'].values, color='blue', label='Santa Barbara MSL')
plt.plot(stas[1]['datestr'].values, stas[1]['msl'].values, color='red', label='Monterey MSL')
plt.xlabel('Time (YYYY/mm/dd)')
plt.ylabel('MSL (m)')
plt.xticks(rotation=45)
plt.legend()
plt.title('Mean Sea Level at Santa Barbara and Monterey, CA')

output_dir = '/Users/kboothomefolder/phy577_projects/hw4/figures'
plt.savefig(os.path.join(output_dir, 'MSL_SB_M.png'))
```

```python
######################## Calculate Statistics ##########################

m0 = round(np.mean(stas[0]['msl']), 3) # sample mean
m1 = round(np.mean(stas[1]['msl']), 3) # sample mean

stdev0 = round(np.std(stas[0]['msl']),2) # standard deviation
stdev1 = round(np.std(stas[1]['msl']),2) # standard deviation

skew0 = round(skew(stas[0]['msl']),3) # skewness
skew1 = round(skew(stas[1]['msl']),3) # skewness

var0 = round(var(stas[0]['msl']),3)
var1 = round(var(stas[1]['msl']),3)

print("Stat Calcs MSL")
print("Santa Barbara Mean:", m0, "m")
print("Monterey Mean:", m1, "m")

print("Santa Barbara SD:", stdev0, "m")
print("Monterey SD:", stdev1, "m")

#print("Santa Barbara Skew:", skew0, "m")
#print("Monterey Skew:", skew1, "m")
```

```python
# Zero center and normalize your records by their respective mean and standard deviation
    
# Zero-Centering: You subtract the mean from each data point in your dataset. 
# This operation shifts the data distribution so that the new mean is 0. 
# This is also referred to as "centering" the data.

# Normalization: You divide each data point by the standard deviation. 
# This operation scales the data so that the new standard deviation is 1.

z0 = (stas[0]['msl'] - m0) / stdev0
z1 = (stas[1]['msl'] - m1) / stdev1

z_corr = np.corrcoef(z0, z1)[0, 1]
varz0 = var(z0)
varz1 = var(z1)
stdz0 = std(z0)
stdz1 = std(z1)

# Make a line for corr coeff
x = np.linspace(min(z1), max(z1), 100)
y = z_corr * x

print(round(z_corr, 2))

# Scatter plot of z-scores for Santa Barbara (z0) and Monterey (z1), corr coeff slope and label

plt.figure(figsize=(8, 6))
plt.scatter(z1, z0, alpha=0.5)
plt.plot(x, y, 'r--', label=f'Correlation Coefficient: {z_corr:.2f}')
plt.text(0.2, 0.8, f'R = {z_corr:.2f}', transform=plt.gca().transAxes, bbox=dict(facecolor='white', edgecolor='black'))
plt.title('Santa Barbara and Monterey Z-Scores')
plt.xlabel('Monterey Z-Scores')
plt.ylabel('Santa Barbara Z-Scores')
plt.grid(True)


###Plot a dashed line with this slope through your scatter plot.

output_dir = '/Users/kboothomefolder/phy577_projects/hw4/figures'
plt.savefig(os.path.join(output_dir, 'MSL_SB_Zscores.png'))
```

```python
# Estimate the lagged auto-correlation function for both z0 and z1

# Correlation measures the relationship between two variables
# Autocorrelation measures the relationship between a variable's current value and its past values.

# Limit the maximum lag to M∆t < 5 days, and time increment (dt)

M = 4 # maximum lag = 4 days
dt = 1 # time interval between data points = 1 day
N = len(z0) # same length for both datasets

# Calculates the number of time lags you need to consider when estimating the auto-correlation function 
# based on the max lag and time interval

Tlag = np.arange(-M, M + 1) * dt

# Make arrays to store the AC values
AC0, AC1 = np.zeros(2 * M + 1), np.zeros(2 * M + 1)

# Calculate the autocorrelation for each lag (m) using loop

for m in range(-M, M + 1):
    AC0[m + M] = np.correlate(z0, np.roll(z0, -m)) / (N * varz0)
    AC1[m + M] = np.correlate(z1, np.roll(z1, -m)) / (N * varz1)

############### Loop Explanation
# calculates the auto-correlation at a lag of m + M for each z-array 
# normalizes it by dividing by the product of the number of data points (N) and the variance of the array 
# Normalization ensures that the auto-correlation values are between -1 and 1 and are independent of the scale 
# and number of data points in each z-array

# np.correlate(z0, np.roll(z0, -m)) 
# calculates the cross-correlation between the original array z0 and a shifted version of z0
# where it is rolled (shifted) by -m positions. Computes the cross-correlation at lag m.

#################

# Difference between correlate and corrcoef?? 

# np.correlate is primarily for signal processing and convolution operations between two arrays 
# np.corrcoef is for calculating correlation coefficients between arrays, 
### typically for assessing linear relationships between data sets

################ Plot Auto-Correlation ####################
# Plot them as line plots versus time τ = m∆t with the same color/linetype convention as before (or include a legend). 

plt.figure(figsize=(10, 6))
plt.plot(Tlag, AC0, label='Santa Barbara, CA', linestyle='-', color='blue')
plt.plot(Tlag, AC1, label='Monterey, CA', linestyle='--', color='red')
plt.title('Lagged Auto-Correlation Function - Daily')
plt.xlabel('Time Lag (days)')
plt.ylabel('Auto-Correlation')
plt.grid(True)
plt.legend()

output_dir = '/Users/kboothomefolder/phy577_projects/hw4/figures'
plt.savefig(os.path.join(output_dir, 'MSL_SB_AutoCorr_Daily.png'))
```

```python
# Estimate the lagged auto-correlation function for both z0 and z1

# Limit the maximum lag to M∆t < 24 hours, and time increment (dt)

M_h = 23 # maximum lag = 23 hours
dt_h = 1 # time interval between data points = 1 hour
Tlag_h = np.arange(-M_h, M_h + 1) * dt_h
AC0_h, AC1_h = np.zeros(2 * M_h + 1), np.zeros(2 * M_h + 1)

for m in range(-M_h, M_h + 1):
    AC0_h[m + M_h] = np.correlate(z0, np.roll(z0, -m)) / (N * varz0)
    AC1_h[m + M_h] = np.correlate(z1, np.roll(z1, -m)) / (N * varz1)

plt.figure(figsize=(10, 6))
plt.plot(Tlag_h, AC0_h, label='Santa Barbara, CA', linestyle='-', color='blue')
plt.plot(Tlag_h, AC1_h, label='Monterey, CA', linestyle='--', color='red')
plt.title('Lagged Auto-Correlation Function - Hourly')
plt.xlabel('Time Lag (hours)')
plt.ylabel('Auto-Correlation')
plt.grid(True)
plt.legend()

output_dir = '/Users/kboothomefolder/phy577_projects/hw4/figures'
plt.savefig(os.path.join(output_dir, 'MSL_SB_AutoCorr_Hourly.png'))
```

```python
#################### Lagged Cross correlation function between z0 and z1 #################################

CC = np.zeros(2 * M_h + 1)

for m in range(-M_h, M_h + 1):
    CC[m + M_h] = np.correlate(z0, np.roll(z1, -m)) / (N * stdz0 * stdz1) # only z1 is lagged, z0 is orig

plt.figure(figsize=(10, 6))
plt.plot(Tlag_h, CC, label='Cross-Correlation', linestyle='-', color='green')
plt.title('Cross-Correlation for Z-Scores - Hourly')
plt.xlabel('Time Lag (hours)')
plt.ylabel('Cross-Correlation')
plt.grid(True)
plt.legend()

output_dir = '/Users/kboothomefolder/phy577_projects/hw4/figures'
plt.savefig(os.path.join(output_dir, 'MSL_SB_CrossCorr_Hourly.png'))
```

```python
# Find the maximum cross-correlation and its corresponding time lag
CCmax = np.max(CC)
ind = np.argmax(CC) # argmax returns the index (or indices) of the maximum value in an array or sequence
lag_max = Tlag_h[ind]

print(f"Maximum Cross-Correlation: {CCmax:.2f}")
print(f"Time Lag: {lag_max} hours")
```

```python
# Remake your scatter plot, but now scatter z1(n + m-max) versus z0. 
# Include a dashed line whose slope is max(ρˆ1,2). How does this compare to your zero lagged version? 

# Calculate M
#M = len(z1) - 1

# Calculate z1(n + m-max)
z1_nmax = np.roll(z1, lag_max)
y = CCmax * z1_nmax

# Scatter plot of z-scores for Santa Barbara (z0) and Monterey (z1), corr coeff slope and label

plt.figure(figsize=(8, 6))
plt.scatter(z1_nmax, z0, alpha=0.5)

# Plot a dashed line with the slope of max(ρ^1,2)
plt.plot(z1_nmax, y, 'r--', label=f'Slope = {CCmax:.2f}')

plt.text(0.2, 0.8, f'R = {CCmax:.2f}', transform=plt.gca().transAxes, bbox=dict(facecolor='white', edgecolor='black'))


plt.title('Santa Barbara and Monterey (lagged) Z-Scores')
plt.xlabel('Monterey Z-Scores (lagged)')
plt.ylabel('Santa Barbara Z-Scores')
plt.grid(True)

output_dir = '/Users/kboothomefolder/phy577_projects/hw4/figures'
plt.savefig(os.path.join(output_dir, 'MSL_SB_Zscores_lagged.png'))
```

```python
# The barotropic tide is a shallow water wave that propagates with a speed, 
# where g is Earth’s gravitational acceleration and d is the water depth. 
# Using the time-lag of maximum correlation τmax and the distance ∆x between the two locations, 
# approximate the average depth of the ocean. 

# CCmax = 0.99
# lag_max = 8 hours

convD = 1609.34 # 1 mile in 1609.34 meters
convT = 3600 # 1 hour = 3600 sec

g = 9.81 # m/s^2
dx = dist * convD # miles --> meters 
t = lag_max * convT # hours --> seconds

# Calculate the wave speed in meters/sec given distance and time lag
c = dx / t # m/sec

# Calculate the average depth of the ocean (d)
dep = c**2 / g

# Print the results
print(f"Wave Speed (c):", round(c, 2), "m/s")
print(f"Average Depth of the Ocean (d):", round(dep, 2),  "meters")
```
