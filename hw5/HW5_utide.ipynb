```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylib import *
from datetime import datetime
import matplotlib.dates as mdates
from scipy.stats import skew
from matplotlib.offsetbox import AnchoredText

# If needed, run this in your folder

#/opt/homebrew/opt/python@3.8/bin/python3.8 -m pip install git+https://github.com/moflaher/ttide_py.git

#import ttide as tt
#from ttide import *
#!pip install ttide

#!pip install folium
#!pip install geopy

import folium
from geopy.distance import geodesic

#################################### Download utide package #######################################

### Not working?
#/opt/homebrew/opt/python@3.8/bin/python3.8 -m pip install https://github.com/wesleybowman/UTide.git

### Not working?
#!pip install utide
#from utide import solve, reconstruct

### Success
import sys  
!{sys.executable} -m pip install --user utide

import utide


##################################### Station Info & Import ###########################################3

# Santa Barbara, CA - Station ID: 9411340 ----- 0
# Monterey, CA - Station ID: 9413450 ------- 1
# metric, GMT, MSL, csv, 6-minute intervals


sta_names = ['Santa Barbara', 'Monterey']

SB_coords = (34.405 , -119.6917)
M_coords = (36.6002, -121.8947)
sta_coords = [SB_coords, M_coords]
print(sta_coords[0])

file_loc = ['/Users/kboothomefolder/phy577_projects/hw5/data/CO-OPS__9411340.csv','/Users/kboothomefolder/phy577_projects/hw5/data/CO-OPS__9413450.csv']
stas = [pd.read_csv(file) for file in file_loc]

##################################### Create Dataframe ###########################################3
#print(stas[0].columns)

stas[0], stas[1] = pd.DataFrame(stas[0].iloc[:, :-5]), pd.DataFrame(stas[1].iloc[:, :-5])

# Convert "Date Time" column to numpy datetime64 array

for i in range(len(stas)):
    stas[i]['datestr'] = np.asarray(pd.to_datetime(stas[i]['Date Time']))
    stas[i]['msl'] = np.asarray(pd.to_numeric(stas[i][' Water Level']))
    stas[i]['sigma'] = np.asarray(pd.to_numeric(stas[i][' Sigma']))
    stas[i].drop(stas[i].columns[0:3], axis=1, inplace=True)
    stas[i]['dt'] = np.asarray(stas[i]['msl'] - mean(stas[i]['msl'])) # Remove the sample mean/detrend
    
##################################### Calculate Statistics ###########################################3

# Create a blank list for storing DataFrames
stats = []

for i in range(len(stas)):
    # Create a dictionary for each calculation
    stat_values = {
        'mean': np.mean(stas[i]['msl']),
        'std': np.std(stas[i]['msl']),
        #'skew': np.skew(stas[i]['msl']),
        'var': np.var(stas[i]['msl'])
    }
    
    # Append the dictionary to the stats list
    stats.append(stat_values)
    print("Stat Calcs for ", sta_names[i])
    print(stats[i])

# stats ---> list of DataFrames containing statistics for each DataFrame in stas


######################################## Tidal Analysis ###############################################

# Perform tidal analysis for DT SB/M MSL
## use solve function to obtain the coefficients

tides = []

for i in range(len(stas)):
    coef = utide.solve(stas[i]['datestr'], stas[i]['dt'], lat=sta_coords[i][0])
    tide = utide.reconstruct(stas[i]['datestr'], coef)
    
    # Define and append to dataframe
    tidal_analysis = {'coef': coef, 'tide': tide}
    tides.append(tidal_analysis)


# print keys
# for analysis in tides:
print("Tidal Analyses:", analysis.keys())
print("Coef/Solve Keys", coef.keys())
print("Reconstruct/Tide Keys", tide.keys())

    
### The amplitudes and phases from the fit are now in the coef data structure (a Bunch), 
### which can be used directly in the reconstruct function to generate a hindcast or 
### forecast of the tides at the times specified in the time array.

### Coef/solve keys include --> dict_keys(['name', 'aux', 'nR', 'nNR', 'nI', 'weights', 'A', 'g', 'mean', 
# 'slope', 'g_ci', 'A_ci', 'diagn', 'PE', 'SNR'])

# The output from the reconstruction is also a Bunch:

# Reconstruct/Tide keys include --> dict_keys(['t_in', 'epoch', 'constit', 'min_SNR', 'min_PE', 't_mpl', 'h'])

########################################### Residual Error calc ######################################################

# Estimate the residual error εp = η1 − ηp and the variance of the error. What is the skill-score of the fit?

# Residual error is the difference between the observed data (n1) and the reconstructed tidal signal (np)
# Variance of the error -- a measure of error value variation

# Calculate the residual error observed - tidal analysis

tide_err = []

for i in range(len(stas)):
    error = stas[i]['dt'] - tides[i]['tide']['h'] # residual error
    var = np.var(error) # residual error variation
    ss = 1 - (var/stats[i]['var']) # skill score

    error_analysis = {'residual error': error, 'error variance': var, 'skill score':ss}
    tide_err.append(error_analysis)
    #print(sta_names[i],f"Residual Error: {round(tide_err[i]['residual error'],3)}")
    print(sta_names[i],f"Error Variance: {round(tide_err[i]['error variance'],3)}")
    print(sta_names[i],f"Skill Score: {round(tide_err[i]['skill score'],3)}")



```
