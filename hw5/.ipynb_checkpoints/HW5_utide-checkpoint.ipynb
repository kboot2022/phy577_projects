```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylib import *
from datetime import datetime
import matplotlib.dates as mdates
from scipy.stats import skew
from matplotlib.offsetbox import AnchoredText
from datetime import timedelta
from scipy.stats import t
import folium
from geopy.distance import geodesic

#################################### Download utide package #######################################

### Not working?
#/opt/homebrew/opt/python@3.8/bin/python3.8 -m pip install https://github.com/wesleybowman/UTide.git

### Not working?
#!pip install utide
#from utide import solve, reconstruct

### Success
import sys  
!{sys.executable} -m pip install --user utide

import utide


##################################### Station Info & Import ###########################################3

# Santa Barbara, CA - Station ID: 9411340 ----- 0
# Monterey, CA - Station ID: 9413450 ------- 1
# metric, GMT, MSL, csv, 6-minute intervals


sta_names = ['Santa Barbara', 'Monterey']

SB_coords = (34.405 , -119.6917)
M_coords = (36.6002, -121.8947)
sta_coords = [SB_coords, M_coords]
#print(sta_coords[0])

file_loc = ['/Users/kboothomefolder/phy577_projects/hw5/data/CO-OPS__9411340.csv','/Users/kboothomefolder/phy577_projects/hw5/data/CO-OPS__9413450.csv']
stas = [pd.read_csv(file) for file in file_loc]

##################################### Create Dataframe ###########################################3
#print(stas[0].columns)

stas[0], stas[1] = pd.DataFrame(stas[0].iloc[:, :-5]), pd.DataFrame(stas[1].iloc[:, :-5])

# Convert "Date Time" column to numpy datetime64 array

for i in range(len(stas)):
    stas[i]['datestr'] = np.asarray(pd.to_datetime(stas[i]['Date Time']))
    stas[i]['msl'] = np.asarray(pd.to_numeric(stas[i][' Water Level']))
    stas[i]['sigma'] = np.asarray(pd.to_numeric(stas[i][' Sigma']))
    stas[i].drop(stas[i].columns[0:3], axis=1, inplace=True)
    stas[i]['dt'] = np.asarray(stas[i]['msl'] - mean(stas[i]['msl'])) # Remove the sample mean/detrend
    
##################################### Calculate Statistics for Orig Dataset ###########################################3

# Create a blank list for storing DataFrames
stats = []

for i in range(len(stas)):
    # Create a dictionary for each calculation
    stat_values = {
        'mean': np.mean(stas[i]['msl']),
        'std': np.std(stas[i]['msl']),
        #'skew': np.skew(stas[i]['msl']),
        'var': np.var(stas[i]['msl'])
    }
    
    # Append the dictionary to the stats list
    stats.append(stat_values)
    #print("Stat Calcs for ", sta_names[i])
    #print(stats[i])

# stats ---> list of DataFrames containing statistics for each DataFrame in stas


######################################## Tidal Analysis ###############################################

# Perform tidal analysis for DT SB/M MSL
## use solve function to obtain the coefficients

t0 = np.array(stas[0]['datestr'])
SB = np.array(stas[0]['dt'])

t1 = np.array(stas[1]['datestr'])
M = np.array(stas[1]['dt'])

#print(SB)


coef_SB = utide.solve(t0, SB, lat=sta_coords[0][0], 
            nodal=False, 
            trend=False, 
            method='ols', 
            conf_int='linear', 
            Rayleigh_min=0.95,)

coef_M = utide.solve(t1, M, lat=sta_coords[1][0], 
            nodal=False, 
            trend=False, 
            method='ols', 
            conf_int='linear', 
            Rayleigh_min=0.95,)

#print(coef_M.keys)

### The amplitudes and phases from the fit are now in the coef data structure (a Bunch), 
### which can be used directly in the reconstruct function to generate a hindcast or 
### forecast of the tides at the times specified in the time array.

### Coef/solve keys include --> dict_keys(['name', 'aux', 'nR', 'nNR', 'nI', 'weights', 'A', 'g', 'mean', 
# 'slope', 'g_ci', 'A_ci', 'diagn', 'PE', 'SNR'])

# The output from the reconstruction is also a Bunch:

# Reconstruct/Tide keys include --> dict_keys(['t_in', 'epoch', 'constit', 'min_SNR', 'min_PE', 't_mpl', 'h'])

## The amplitudes and phases from the fit are now in the coef data structure (a Bunch), 
## which can be used directly in the reconstruct function to generate a hindcast or forecast of the tides 
## at the times specified in the time array.

tide_SB = utide.reconstruct(stas[0]['datestr'], coef_SB, verbose=False)
tide_M = utide.reconstruct(stas[0]['datestr'], coef_M, verbose=False)


# Cut this out because possibly incorrect?

############################## Estimate Residual Error for utide analysis #################################

################################# Using "tides" ##################################################

# Santa Barbara 

tide_err = []

for i in range(len(stas)):
    SB_error = stas[0]['dt'] - tide_SB['h'] # residual error
    SB_var = np.var(SB_error) # residual error variation
    SB_ss = 1 - (SB_var/stats[0]['var']) # skill score

    SB_err_analysis = {'residual error': SB_error, 'error variance': SB_var, 'skill score':SB_ss}
    tide_err.append(SB_err_analysis)

    M_error = stas[1]['dt'] - tide_M['h'] # residual error
    M_var = np.var(M_error) # residual error variation
    M_ss = 1 - (M_var/stats[1]['var']) # skill score

    M_err_analysis = {'residual error': M_error, 'error variance': M_var, 'skill score':M_ss}
    
    tide_err.append(M_err_analysis)
    
    #print(sta_names[0],f"Residual Error: {constit} {round(tide_err[0]['residual error'],3)}")
    #print(sta_names[i],f"Error Variance: {round(tide_err[0]['error variance'],3)}")
    #print(sta_names[i],f"Skill Score: {round(tide_err[0]['skill score'],3)}")

print('Santa Barbara Error Variance: ', round(tide_err[0]['error variance'],4)) 
print('Santa Barbara Skill Score: ', round(tide_err[0]['skill score'],4)) 
print('Monterey Error Variance: ', round(tide_err[1]['error variance'],4)) 
print('Monterey Skill Score: ', round(tide_err[1]['skill score'],4)) 

################################### Reconstruct tidal constituents ##########################################

tidal_constituents = coef_SB['name']

tide_SB_con = []

for constituent in tidal_constituents:
    con_tide = utide.reconstruct(t0, coef_SB, constit=[constituent], verbose=False)
    tide_SB_con.append(con_tide)

tide_M_con = []

for constituent in tidal_constituents:
    con_tide = utide.reconstruct(t1, coef_M, constit=[constituent], verbose=False)
    tide_M_con.append(con_tide)


############################## Estimate Residual Error for tidal constituent reconstruction #################################

# Santa Barbara 

# Calculate the amplitude, error, and amplitude-to-error ratio for each constituent

A_SB = coef_SB['A']
A_M = coef_M['A']
err_SB = coef_SB['g_ci'][0]
err_M = coef_M['g_ci'][0]
    
ratio_SB = A_SB/err_SB
ratio_M = A_M/err_M

sig_SB = []
sig_M = []
val = 1

for i in range(len(A_SB)):

    if ratio_SB[i] > val:  
        sig_SB.append(tidal_constituents[i])
    if ratio_M[i] > val:  
        sig_M.append(tidal_constituents[i])

print("SB Significant Tidal Constituents: ", sig_SB)
#print(ratio_SB)
print("M Significant Tidal Constituents: ", sig_M)  
#print(ratio_M)

##################################### Determine Best Fit Coeffiecients ################################################################

# Find the best fit coefficients (a) using least-squares function equivalent in Python

t0 = np.array(stas[0]['datestr'])
t1 = np.array(stas[1]['datestr'])

######## Convert time record to hours

# Calculate total minutes passed since the first observation
#mins = (t0 - t0[0]).astype('timedelta64[m]').astype(float)

mins = (t0 - t0[0]).astype('timedelta64[m]').astype(float)
hours = np.round((mins / 60.0), decimals=5)
days = np.round((hours/24), decimals = 5)

#print("Days: ", days)

# Tidal Frequencies

M2_om = 2*pi/12.42; # lunar semidiurnal consitiuent
S2_om = 2*pi/12;  # solar semidiurnal constituent

# Make unit amplitude functions

cos_M2 = cos(M2_om * hours);
sin_M2 = sin(M2_om * hours);
cos_S2 = cos(S2_om * hours);
sin_S2 = sin(S2_om * hours);

# Concatenating the arrays vertically to create matrix H

H = np.column_stack((cos_M2, sin_M2, cos_S2, sin_S2))

#print(H)
#print(H.shape)  # (7440, 4)

# Find best fit coefficients using detrended water level(SB)

#### What is equivalent to MATLAB’s built in least-squares function ‘\’,
### Matlab edition --> a = H \ SB

# Solve for 'a' using least-squares solution ### See notes below for details on Python least sqares solution

a, residuals, rank, s = np.linalg.lstsq(H, SB, rcond=None)

print("Solution for 'a':", a)

# Compute the inverse covariance matrix
### MAtlab edition --> inv_cov_matrix = inv(H'*H)

# Calculate the transpose of H
H_p = np.transpose(H)

# Compute H_transpose multiplied by H

calc = np.dot(H_p, H)  

# Calculate the inverse of the result
inv_cov_matrix = np.linalg.inv(calc)

print("SB Inverse Covariance Matrix: ", inv_cov_matrix)

'''
The MATLAB backslash operator (\) is used to solve systems of linear equations, and it's often used 
for least-squares regression when dealing with overdetermined systems. 
In Python, specifically with NumPy, you can achieve similar functionality using numpy.linalg.lstsq().

'''

############################################ Fourier Series ########################################################

# Estimate the residual error εˆ = η1 − H ˆa and the variance of the error s2. 
# What is the skill-score of the fit SSηˆ?

# Perform matrix-vector multiplication: Ha = H*a

Ha = np.dot(H, a)

# Calculate the residual error

res_e = SB - Ha

res_err_var = np.var(res_e)

# Skill score

# calculate variance of error of original data

SB_var = np.var(SB)

#HSB_ss = 1-(SB_var/res_err_var)

HSB_ss = 1-(res_err_var/SB_var)

print("LSF residual error variance:", round(res_err_var, 4))
print("LSF skill score:", round(HSB_ss, 4))

################### LSF - amplitude and phase of M2 and S2 fits ########################3

# extract coefficients

M2_a = a[0]
M2_b = a[1]
S2_a = a[2]
S2_b = a[3]

# Calculate amplitude and phase for both constituents

# *******  These are for meters squared, when should be meters only, for later we do the square root of the sum
M2_amp = (M2_a ** 2)+(M2_b ** 2)
M2_pha = math.atan2(M2_b, M2_a)

S2_amp = (S2_a ** 2)+(S2_b ** 2)
S2_pha = math.atan2(S2_b, S2_a)

print('M2 amplitude:', round(M2_amp,3), ' M2 Phase:', round(M2_pha,3))
print('S2 amplitude:', round(S2_amp,3), ' S2 Phase:', round(S2_pha,3))

# Fraction of total variance of LSF explained by M2 and S2

M2_frac = M2_amp / (2 * SB_var)
S2_frac = S2_amp / (2 * SB_var)

print('Fraction of total variance of LSF explained by M2 and S2')
print('M2 Fraction:', round(M2_frac, 3))
print('S2 Fraction:', round(S2_frac, 3))

# Difference in amplitude between LSF and utide

LSF_amp = M2_amp + S2_amp
utide_amp = sum(A_SB)

#print('LSF', LSF_amp)
#print('Utide', (A_SB[0] + A_SB[1])) # for all constituents

amp_dif = 100 * ((abs(utide_amp-LSF_amp))/(0.5*(utide_amp+LSF_amp)))

print('Difference in amplitude between LSF and Utide: ', round(amp_dif), '%')

#################################### Auto Correlation ################################################

# We can approximate the error for our fit amplitudes using the student-t distribution 
# assuming ν = Neff − 4 degrees of freedom. To compute Neff = N/Nd we first need to 
# estimate the decorellation scale Nd. Estimate the lagged auto-correlation for both 
# errors εˆand εp out to 15 days (do not compute the cross-correlation). 
# Plot the auto-correlation versus lag and indicate the integral time-scale for both with a 
# vertical line. What are Nd and Nef f for each?

# estimate lagged auto-correlation for both errors out to 15 days 

# N --> # of observations
# Neff --> N/Nd where Nd is decorrelation scale

N = len(t0)

# Estimate the decorrelation scale

dt = days[1] - days[0] # change in time for each step -- 6 mins

# calculate number of occurrences of dt within a total of 15 days

Mv = round(15 / dt)

lags = np.arange(0, Mv + 1)

utide_R = np.empty(Mv + 1)
LSF_R = np.empty(Mv + 1)

utide_SB_err = np.asarray(tide_err[0]['residual error'])
LSF_err_SB = np.asarray(res_e)

#print('Utide:', len(utide_SB_err)) # problem with utide, is scalar
#print('LSF:', len(LSF_err_SB)) # is numpy array


for m in range(Mv + 1):
    ind = np.arange(0, N - m)
    utide_R[m] = np.sum(utide_SB_err[ind + m] * utide_SB_err[ind]) / N
    LSF_R[m] = np.sum(LSF_err_SB[ind + m] * LSF_err_SB[ind]) / N   

utide_R = utide_R / utide_R[0]
LSF_R = LSF_R / LSF_R[1]


# calculate Nd and Neff for each error
# Nd --> where autocrrelation drops to 1/e of initial value
# Neff --> N/Nd where Nd is decorrelation scale

Nd_utide = np.where(utide_R <= 1/np.exp(1))[0][0]
Nd_LSF = np.where(LSF_R <= 1/np.exp(1))[0][0]
Neff_utide = N / Nd_utide
Neff_LSF = N / Nd_LSF

print('Utide Nd:', Nd_utide, 'Neff:', round(Neff_utide,2))
print('LSF Nd:', Nd_LSF, 'Neff:', round(Neff_LSF,2))

################### Find ratio of confidence limits of LSF and compare to utide #########################3

ci = 0.05 # confidence interval
v = Neff_LSF - 4 # v = Neff - 4 degrees of freedom

# Calculate the t-score using the inverse CDF of the t-distribution

t_score = t.ppf(1 - ci/2, v)
s_eps = sqrt(res_err_var) # residual error variance for LSF 
C_inv = inv_cov_matrix[0, 0] # diagonal element of C^(-1)
con_int = t_score * sqrt(s_eps * C_inv)

# Lower and upper bounds of confidence interval
low = M2_amp - con_int
up = M2_amp + con_int

print("Lower LSF CI:", round(low, 3))
print("Upper LSF CI:", round(up, 3))
```

```python
############################## Solve for Tidal Constituents ###########################################################

#tidal_constituents = coef_SB['name']
print(tidal_constituents)

M_con_tides = {}
SB_con_tides = {}

for constituent in tidal_constituents:
    
    # Utide solve - tidal analysis for each specific constituent (characterize)
    
    SB_con_coef = utide.solve(t0, SB, lat=sta_coords[0][0], 
            constit=[constituent], 
            nodal=False, 
            trend=False, 
            method='ols', 
            conf_int='linear', 
            Rayleigh_min=0.95,)
    
    M_con_coef = utide.solve(t1, M, lat=sta_coords[1][0], 
            constit=[constituent], 
            nodal=False, 
            trend=False, 
            method='ols', 
            conf_int='linear', 
            Rayleigh_min=0.95,)
    
    # Utide reconstruct - tidal signal for specific constituents (fit & forecast)

    tide_SB_con = utide.reconstruct(t0, SB_con_coef, constit=[constituent])
    tide_M_con = utide.reconstruct(t1, M_con_coef, constit=[constituent])


    # Store the reconstructed tidal signal in the dictionary
    SB_con_tides[constituent] = tide_SB_con['h']
    M_con_tides[constituent] = tide_M_con['h']



########################################### Identify Sig Cons with Skill Score ###############################################

#If SS is close to 1, it means that the variance of the residual errors is small compared to the variance of the original data, 
#indicating a good performance of the model

# set the skill score threshold

sst = 0.1

# Use a for-loop to find the skill score and residual error for each tidal constituent

SB_outputs = {}
M_outputs = {}

# Iterate over each constituent

for constituent in tidal_constituents:

    res_err_SB = SB - tide_SB_con['h']
    res_err_M = M - tide_M_con['h']

    err_var_SB = np.var(res_err_SB)
    err_var_M = np.var(res_err_M)

    orig_var_SB = np.var(SB)
    orig_var_M = np.var(M)

    ss_SB = 1 - (err_var_SB / orig_var_SB)
    ss_M = 1 - (err_var_M / orig_var_M)

    SB_outputs[constituent] = {
        'Residual Error': res_err_SB,
        'Variance of Error': err_var_SB,
        'Original Data Variance': orig_var_SB,
        'Skill Score': ss_SB
    }
    
    M_outputs[constituent] = {
        'Residual Error': res_err_M,
        'Variance of Error': err_var_M,
        'Original Data Variance': orig_var_M,
        'Skill Score': ss_M
    }
    
    '''print(f"Santa Barbara \nConstituent: {constituent}")
    print(f"Residual Error: {res_err_M}")
    print(f"Variance of Error: {err_var_M:.6f}")
    print(f"Original Data Variance: {orig_var_M:.6f}")
    print(f"Skill Score: {ss_SB:.6f}")
    print(f"Skill Score - Monterey: {ss_M:.6f}")'''



'''

most_significant_constituents = [constituent for constituent, results in results_dict.items() if results['Skill Score'] > threshold]

print(f"\nMost Significant Constituents based on Skill Score: {most_significant_constituents}")

'''
```
